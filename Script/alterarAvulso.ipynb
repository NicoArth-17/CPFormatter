```python
import pandas as pd
from itertools import zip_longest

df = pd.read_excel('arquivos/cpf unicred.xlsx')

display(df)

cpfs = df.iloc[:,0]

cpfs_formatados = []

cpfinvalido = []


for cpf in cpfs:

    cpf = str(cpf)

    if len(cpf) == 11: 
        cpf = f'{cpf[:3]}.{cpf[3:6]}.{cpf[6:9]}-{cpf[9:]}'
        cpfs_formatados.append(cpf)

    elif len(cpf) == 10:
        cpf = f'0{cpf[:2]}.{cpf[2:5]}.{cpf[5:8]}-{cpf[8:]}'
        cpfs_formatados.append(cpf)

    elif len(cpf) == 9:
        cpf = f'00{cpf[:1]}.{cpf[1:4]}.{cpf[4:7]}-{cpf[7:]}'
        cpfs_formatados.append(cpf)

    elif len(cpf) == 8:
        cpf = f'000.{cpf[:3]}.{cpf[3:6]}-{cpf[6:]}'
        cpfs_formatados.append(cpf)

    elif len(cpf) == 7:
        cpf = f'000.0{cpf[:2]}.{cpf[2:5]}-{cpf[5:]}'
        cpfs_formatados.append(cpf)

    else:
        cpfinvalido.append(cpf)
        cpfs_formatados.append(cpf)

print(cpfs_formatados)
print(cpfinvalido)


new_df = pd.DataFrame(list(zip_longest(cpfs_formatados, cpfinvalido)), columns=['CPF', 'CPF Inválido'])
new_df.to_excel('arquivos/Formatted CPFs.xlsx', index=None)
```
